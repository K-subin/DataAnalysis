```
import torch
import requests
import matplotlib.pyplot as plt
```

```
# iris data 다운로드 받기
iris_url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
r = requests.get(iris_url)

with open("iris.data","wb") as f:
  f.write(r.content)
```

```
# iris data 불러오기 및 데이터 포맷 변환
vectors = []
answers = []
with open("iris.data","r") as f:
  for line in f:
    if len(line.strip()) == 0: continue
    items = line.strip().split(",")
    vectors.append([float(i) for i in items[:4]])
    answers.append(items[4])
```

```
# 모든 축에 대한 평균 구해서 평균을 기준으로 축 옮기기

X = torch.FloatTensor(vectors)
Z = X - X.mean(axis=0)

# Z에서 데이터에서 주성분 찾기

def findPC(Z):
  w = torch.randn(Z.shape[1]) # 랜덤
  w = w / (w ** 2).sum() ** 0.5 # 단위행렬 - 사이즈 1 되도록

  lr = 0.1

  for epoch in range(1001):
    w.requires_grad_(True)
    variance = ((Z * w).sum(axis=1) ** 2).mean() # 분산

    variance.backward()
    with torch.no_grad():
      w = w + lr * w.grad # 경사를 증가시키기 위해
      w = w / (w ** 2).sum() ** 0.5

  return w, variance.item()

# 내적 * w
# 4차원 데이터니까 4번까지 의미있다.
def PCA(Z, n):
  W, V = [], []
  for _ in range(n):
    w, v = findPC(Z)
    W.append(w.tolist())
    V.append(v)
    Z = Z - (Z * w).sum(axis=1).view(-1,1) * w
  
  return W, V

def transform(Z, W):
  return Z @ torch.Tensor(W).T #torch.mm(Z, torch.Tensor(W).T)

W, V = PCA(Z, 4)

plt.title('score plot')
plt.plot(range(1, len(V)+1), V, "-o")
plt.show()

# 각 축이 주성분이 되도록 기존 데이터를 변경
K = transform(Z,W)

species = {s:i for i,s in enumerate(set(answers))}

plt.scatter(K[:,0], K[:,1], c=[species[a] for a in answers], alpha=0.3)
plt.show()
```
